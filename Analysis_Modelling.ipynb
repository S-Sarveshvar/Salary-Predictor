```python
import pandas as pd
```

```python
data = pd.read_excel("Employees.xlsx")
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

# Count values in the Gender column and plot as pie chart
data["Gender"].value_counts().sort_values(ascending=False).plot(kind="pie", autopct='%1.1f%%')

# Add title and remove y-axis label
plt.title("Pie Chart of Gender Column")
plt.ylabel("")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.hist(data["Job Rate"], bins=10, edgecolor='black')  # You can adjust `bins`
plt.title("Histogram of Job Rates")
plt.xlabel("Rate")
plt.ylabel("Count")
plt.show()
```

```python
data["Job Rate"].describe()
```

```python
data.groupby("Department")["Annual Salary"].mean().sort_values(ascending=False).head(7).plot(kind="bar")

plt.title("Average Salary by Department (Top 7)")
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

data.groupby("Country")["Job Rate"].mean().sort_values(ascending=False).plot(kind="bar", figsize=(10, 5))

plt.title("Average Job Rates by Country")
plt.xlabel("Country")
plt.ylabel("Job Rate Average")
plt.xticks(rotation=45)  # Optional: makes x-labels readable
plt.tight_layout()
plt.show()
```

```python
plt.hist(data["Overtime Hours"], bins=10, edgecolor='black')
plt.title("Histogram of Overtime Hours")
plt.xlabel("Overtime Hours")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
X = data[["Years", "Job Rate"]]
y = data["Annual Salary"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
len(y_train)  # should be ~80% of data
#len(X_test)   # should be ~20% of data
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
predslr = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
import joblib
joblib.dump(lr, "linearmodel.pkl")
```

mean_absolute_error(predslr, y_test)
